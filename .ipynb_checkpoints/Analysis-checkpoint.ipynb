```python
timeStamp = data_set["Timestamp"]
views = data_set["14. Have your views on abortion changed over time?"]
data = {
    "Timestamp": timeStamp,
    # ... (other columns)
    "Have your views on abortion changed over time?": views
                                                     
}
dataframes = pd.DataFrame(data)
dataframes
```

```python
dataframes.head(20)
```

```python
sum_of_zeros = (dataframes["Have your views on abortion changed over time?"] == 0).sum()
sum_of_ones = (dataframes["Have your views on abortion changed over time?"] == 1).sum()
sum_of_twos = (dataframes["Have your views on abortion changed over time?"] == 2).sum()

print(f"yes have become more pro-life: {sum_of_zeros}")
print(f"yes have become more pro-choice: {sum_of_ones}")
print(f"no my views have stayed the same: {sum_of_twos}")


analysis_on_views = dataframes["Have your views on abortion changed over time?"]

# Calculate mean and standard deviation
mean_value = analysis_on_views.mean()
std_deviation = analysis_on_views.std()
statistical_summary = analysis_on_views.describe()

print(f"Mean: {mean_value:.2f}")
print(f"Standard Deviation: {std_deviation:.2f}")
print("Statistical Summary:")
print(statistical_summary)

```

```python
# Applying t_statistics and t_test on the responses from question  8 and 9

pro_test = data_set['8.  Are you familiar with the terms “pro-life” and “pro-choice”?']
pro_test2 = data_set['9. What is your stance on abortion?']
from scipy.stats import ttest_rel
t_statistics, p_value = ttest_rel(pro_test, pro_test2)
print("t_statistics:", str(t_statistics))
print("p_value", str(p_value))
```

```python
## finding the regression between the responses form 8 and 9
from scipy.stats import linregress, chisquare
regression_result = linregress(pro_test, pro_test2)
sum1_test = pro_test.sum()
sum2_test = pro_test2.sum()

print("The regression between the responses is", str(regression_result))
```

```python
arr = np.array(pro_test)
arr2 = np.array(pro_test2)
chisquares_1 =  chisquare(arr)
chi_square_2 = chisquare(arr2)
print("chisquares value for '8.  Are you familiar with the terms “pro-life” and “pro-choice”?'", str(chisquares_1[0]))
print("chisquares value for  '9. What is your stance on abortion?'", str(chi_square_2[0]))
```
